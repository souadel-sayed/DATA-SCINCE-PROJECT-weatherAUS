```python
t = (df.dtypes == "float64")
num_cols = list(t[t].index)

print("Neumeric variables:")
print(num_cols)
```

```python
for i in num_cols:
    df[i].fillna(df[i].median(), inplace=True)
    
df.info()
```

```python
o = (df.dtypes == "object")
obj_cols = list(t[o].index)

print("object  variables:")
print(obj_cols)
```

```python
for i in obj_cols:
    df[i].fillna(df[i].mode()[0], inplace=True)
    
df.info()
```

```python
df.drop("Date",axis=1,inplace=True)
```

```python
df['RainToday'].replace({'No': 0, 'Yes': 1},inplace = True)
df['RainTomorrow'].replace({'No': 0, 'Yes': 1},inplace = True)
```

```python
df.shape
```

```python
df.head()
```

```python
df = pd.concat(
                     [df
                     ,pd.get_dummies(df.Location), 
                     pd.get_dummies(df.WindGustDir),
                     pd.get_dummies(df.WindDir9am),
                     pd.get_dummies(df.WindDir3pm)], axis=1)
```

```python
df.head()
```

```python
df.drop(["WindGustDir","WindDir3pm","WindDir9am","Location"],axis=1,inplace=True)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
y=df.RainTomorrow
x=df.drop("RainTomorrow",axis=1)
```

```python
y.shape
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=45)
```

```python
X_train
```

```python
X_train.astype(int)
```

```python
X_test.astype(int)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
# instantiate the model
logreg = LogisticRegression(solver='liblinear', random_state=0)
# fit the model
logreg.fit(X_train, y_train)
```

```python
logreg.score(X_train, y_train)
```

```python
y_pred_test = logreg.predict(X_test)

y_pred_test
```

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred_test)))
```

```python
from sklearn.tree import DecisionTreeClassifier

params_dt = {'max_depth': 16,
             'max_features': "sqrt"}

model_dt = DecisionTreeClassifier(**params_dt)
```

```python
model_dt.fit(X_train, y_train)
```

```python
model_dt.score(X_train, y_train)
```

```python
y_pred_test = model_dt.predict(X_test)

y_pred_test
```

```python
print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred_test)))
```
